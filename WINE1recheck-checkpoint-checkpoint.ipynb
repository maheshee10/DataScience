```python
x = wine.drop('quality',1)
y = wine.quality
print(x.shape,y.shape)
```

```python
#BestFeatures
```

```python
kbest = SelectKBest(score_func = chi2, k = 11)
best = kbest.fit(x,y)
print(best.scores_)
```

```python
features = kbest.transform(x)
print(features)
features = pd.DataFrame(features, columns = ['fixed acidity','volatile acidity','citric acid','residual sugar', 
                                             'chlorides','free sulfur dioxide','density','pH','sulphates',
                                             'alcohol','Type'])
features
```

```python
x = features
y = wine.quality
print(x.shape,y.shape)
```

```python
#model evaluation
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y, test_size = 0.33, random_state =1)
```

```python
models = []
models.append(('LR', LogisticRegression()))
models.append(('CART',DecisionTreeClassifier()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('SVC',SVC()))
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('NB',GaussianNB()))
names = []
scores = []
for name, model in models:
    kfold = model_selection.ShuffleSplit(n_splits = 5, random_state=1) #ShuffleSplitvalidation
    model.fit(x_train,y_train)
    cv_score = model_selection.cross_val_score(model, x, y,cv = kfold) #cvscore
    pred_y = model.predict(x_test)
    accuracy = accuracy_score(y_test,pred_y)
    names.append(name)
    scores.append(accuracy)
    print(name, 'accuracy', accuracy*100)
    print(pred_y)
    
```

```python
df = pd.DataFrame(scores, names)
df.plot.bar(figsize = (10,6))
```

```python
#Ensembling
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y, test_size = 0.33, random_state =1)
```

```python
models = []
models.append(('RF', RandomForestClassifier()))
models.append(('ADA',AdaBoostClassifier()))
models.append(('XTR',ExtraTreesClassifier()))
models.append(('GB',GradientBoostingClassifier()))
names = []
scores = []
for name, model in models:
    kfold = model_selection.ShuffleSplit(n_splits = 5, random_state=1) #ShuffleSplitvalidation
    model.fit(x_train,y_train)
    cv_score = model_selection.cross_val_score(model, x, y,cv = kfold) #cvscore
    pred_y = model.predict(x_test)
    accuracy = accuracy_score(y_test,pred_y)
    names.append(name)
    scores.append(accuracy)
    print(name, 'accuracy', accuracy*100)
    print(pred_y)
    
```

```python
df = pd.DataFrame(scores,names)
df.plot.bar() #RandomForestperforms well
```
