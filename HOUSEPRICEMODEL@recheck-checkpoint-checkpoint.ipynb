```python
#ModelEvaluation
x_train, x_val, y_train, y_val = model_selection.train_test_split(x,y, test_size = 0.33, random_state =1)
```

```python
models = []
models.append(('LR', LinearRegression()))
models.append(('CART',DecisionTreeRegressor()))
models.append(('NBR',KNeighborsRegressor()))
names = []
rmss = []
for name, model in models:
    kfold = model_selection.ShuffleSplit(n_splits = 5, random_state=1) #ShuffleSplitvalidation
    model.fit(x_train,y_train)
    cv_score = model_selection.cross_val_score(model, x, y,cv = kfold) #cvscore
    print(name, cv_score.mean()*100)
    pred_x_val = model.predict(x_val)
    rms = mean_squared_error(pred_x_val,y_val) #rms
    names.append(name)
    rmss.append(rms)
    msg = '%s %f' % (name, math.sqrt(rms))
    print('/n', msg)
```

```python
df = pd.DataFrame(rmss,names)
df.plot.bar()
```

```python
#LinearRegression model gives better performance
```

```python
#PredictingModel
```

```python
i = 1
kf = model_selection.StratifiedKFold(n_splits = 5, random_state = 1, shuffle = True)
for train_index, test_index in kf.split(x,y) :
    print('\n {} of kfold {}'.format(i,kf.n_splits))
    xtr, xvl = x.loc[train_index], x.loc[test_index]
    ytr, yvl = y.loc[train_index], y.loc[test_index]
    model = LinearRegression()
    model.fit(xtr,ytr)
    pred_xvl = model.predict(xvl)
    rms  = mean_squared_error(pred_xvl,yvl)
    print('RMSE', math.sqrt(rms))
    i+=1
    pred_test = model.predict(test)
```

```python
i = 1
kf = model_selection.StratifiedKFold(n_splits = 5, random_state=1, shuffle= True)
for train_index, test_index in kf.split(x,y) :
    print('\n {} of kfold{}'.format(i,kf.n_splits))
    x_train,x_val = x.loc[train_index], x.loc[test_index]
    y_train, y_val = y.loc[train_index], y.loc[test_index]
    model = RandomForestRegressor()
    model.fit(x_train,y_train)
    pred_xval = model.predict(x_val)
    mse = mean_squared_error(pred_xval,y_val)
    print('RMSE', math.sqrt(mse))
    i+=1
    pred_test = model.predict(test)
    print(pred_test)
```
